# IMPORT DES DONNEES ET LIBRERIES ET DEFINITION FUNCTIONS

###### -------------------------------------------------------------------------

##### J'import des donnèes et je declare certains functions

```python
import warnings
warnings.simplefilter("ignore")
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import datetime
import scipy.stats as st
import statsmodels.api as sm
customers = pd.read_csv ('customers.csv')
products = pd.read_csv ('products.csv')
transactions = pd.read_csv ('transactions.csv')
def gini(arr):
    ## first sort
    sorted_arr = arr.copy()
    sorted_arr.sort()
    n = arr.size
    coef_ = 2. / n
    const_ = (n + 1.) / n
    weighted_sum = sum([(i+1)*yi for i, yi in enumerate(sorted_arr)])
    return coef_*weighted_sum/(sorted_arr.sum()) - const_

def lorenz(arr):
    # this divides the prefix sum by the total sum
    # this ensures all the values are between 0 and 1.0
    scaled_prefix_sum = arr.cumsum() / arr.sum()
    # this prepends the 0 value (because 0% of all people have 0% of all wealth)
    return np.insert(scaled_prefix_sum, 0, 0)
import scipy.stats 
def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
    dfn = x.size-1 #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-scipy.stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
    return f, p
def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
```

```python
ca['ventes_A (€)']=0
ca['ventes_B (€)']=0
ca['ventes_C (€)']=0

ca['ventes_A (€)']=ca['ventes_A (€)'].fillna(0)
ca['ventes_B (€)']=ca['ventes_B (€)'].fillna(0)
ca['ventes_C (€)']=ca['ventes_C (€)'].fillna(0)

ca['ventes_A (€)']=ca['price'][ca['categ']=='A']
ca['ventes_B (€)']=ca['price'][ca['categ']=='B']
ca['ventes_C (€)']=ca['price'][ca['categ']=='C']

cap=ca[['id_prod','categ','price','session_id']].groupby(['id_prod', 'price', 'categ']).count()
cap.rename(columns={'session_id':'ventes_prod'}, inplace=True)
cap=cap.reset_index()
cap['Chiffres d’affaires de produit (€)']=cap['price']*cap['ventes_prod']

cap['CA 0-1']=1*round( (cap['Chiffres d’affaires de produit (€)'] - cap['Chiffres d’affaires de produit (€)'].min() ) / (cap['Chiffres d’affaires de produit (€)'].max() - cap['Chiffres d’affaires de produit (€)'].min() ),4)
cap['VP 0-1']=1*round( (cap['ventes_prod'] - cap['ventes_prod'].min() ) / (cap['ventes_prod'].max() - cap['ventes_prod'].min() ),4)
#------
Aq3, Aq1 = np.percentile (cap['Chiffres d’affaires de produit (€)'][cap['categ']=='A'], [75, 25])
Aiqr = Aq3 - Aq1
Alow=Aq1-(1.5*Aiqr)
Ahigh=Aq3+(1.5*Aiqr)

Bq3, Bq1 = np.percentile (cap['Chiffres d’affaires de produit (€)'][cap['categ']=='B'], [75, 25])
Biqr = Bq3 - Bq1
Blow=Bq1-(1.5*Biqr)
Bhigh=Bq3+(1.5*Biqr)

Cq3, Cq1 = np.percentile (cap['Chiffres d’affaires de produit (€)'][cap['categ']=='C'], [75, 25])
Ciqr = Cq3 - Cq1
Clow=Cq1-(1.5*Ciqr)
Chigh=Cq3+(1.5*Ciqr)


q3, q1 = np.percentile (cap['Chiffres d’affaires de produit (€)'], [75, 25])
iqr = q3 - q1
low=q1-(1.5*iqr)
high=q3+(1.5*iqr)
#------
cap['Outlier / Inliers (CA)']=''
cap['Outlier / Inliers (CA)'][(cap['Chiffres d’affaires de produit (€)']<=high) & (cap['Chiffres d’affaires de produit (€)']>=low)]='Inlier'
cap['Outlier / Inliers (CA)'][ (cap['Chiffres d’affaires de produit (€)']>high) | (cap['Chiffres d’affaires de produit (€)']<low) ]='Outlier'
#------
cap['Outlier / Inliers (CA)(C)']=''
cap['Outlier / Inliers (CA)(C)'][(cap['Chiffres d’affaires de produit (€)']<=Ahigh) & (cap['Chiffres d’affaires de produit (€)']>=Alow) & (cap['categ']=='A' ) ]='Inlier A'
cap['Outlier / Inliers (CA)(C)'][ ((cap['Chiffres d’affaires de produit (€)']>Ahigh) | (cap['Chiffres d’affaires de produit (€)']<Alow)) & (cap['categ']=='A' ) ]='Outlier A'
cap['Outlier / Inliers (CA)(C)'][(cap['Chiffres d’affaires de produit (€)']<=Bhigh) & (cap['Chiffres d’affaires de produit (€)']>=Blow) & (cap['categ']=='B' ) ]='Inlier B'
cap['Outlier / Inliers (CA)(C)'][ ((cap['Chiffres d’affaires de produit (€)']>Bhigh) | (cap['Chiffres d’affaires de produit (€)']<Blow) )& (cap['categ']=='B' ) ]='Outlier B'
cap['Outlier / Inliers (CA)(C)'][(cap['Chiffres d’affaires de produit (€)']<=Chigh) & (cap['Chiffres d’affaires de produit (€)']>=Clow) & (cap['categ']=='C' ) ]='Inlier C'
cap['Outlier / Inliers (CA)(C)'][ ((cap['Chiffres d’affaires de produit (€)']>Chigh) | (cap['Chiffres d’affaires de produit (€)']<Clow)) & (cap['categ']=='C' ) ]='Outlier C'
#------
cac=ca[['client_id','price', 'sex', 'Age','Age (c)','birth','session_id','ventes_A (€)','ventes_B (€)','ventes_C (€)','ventes_A (€) (f)','ventes_B (€) (f)','ventes_C (€) (f)','ventes_A (€) (m)','ventes_B (€) (m)','ventes_C (€) (m)']].groupby(['client_id','sex','Age','Age (c)','birth']).agg({'price':np.sum, 'ventes_A (€)':np.sum,'ventes_B (€)':np.sum,'ventes_C (€)':np.sum,'session_id': "count", 'ventes_A (€) (f)':np.sum,'ventes_B (€) (f)':np.sum,'ventes_C (€) (f)':np.sum, 'ventes_A (€) (m)':np.sum,'ventes_B (€) (m)':np.sum,'ventes_C (€) (m)':np.sum})
cac.rename(columns={'price':'ventes_client (€)'}, inplace=True),cac.rename(columns={'session_id':'nombre_sessions'}, inplace=True)
cac=cac.reset_index()

cas=ca[['session_id','price','date', 'sex', 'Age','Age (c)','birth', 'ventes_A (€)','ventes_B (€)','ventes_C (€)','ventes_A (€) (f)','ventes_B (€) (f)','ventes_C (€) (f)','ventes_A (€) (m)','ventes_B (€) (m)','ventes_C (€) (m)' ]].groupby('session_id').agg({'date':np.mean,'price':np.sum, 'Age':np.mean,'Age (c)':np.max, 'sex':np.max, 'birth':np.mean,'ventes_A (€)':np.sum, 'ventes_B (€)':np.sum, 'ventes_C (€)':np.sum , 'ventes_A (€) (f)':np.sum,'ventes_B (€) (f)':np.sum,'ventes_C (€) (f)':np.sum, 'ventes_A (€) (m)':np.sum,'ventes_B (€) (m)':np.sum,'ventes_C (€) (m)':np.sum})
cas.rename(columns={'ventes_A (€)':'ventes_A_session (€)'}, inplace=True),cas.rename(columns={'ventes_B (€)':'ventes_B_session (€)'}, inplace=True),cas.rename(columns={'ventes_C (€)':'ventes_C_session (€)'}, inplace=True),cas.rename(columns={'ventes_A (€) (f)':'ventes_A_session (€) (f)'}, inplace=True),cas.rename(columns={'ventes_B (€) (f)':'ventes_B_session (€) (f)'}, inplace=True),cas.rename(columns={'ventes_C (€) (f)':'ventes_C_session (€) (f)'}, inplace=True),cas.rename(columns={'ventes_A (€) (m)':'ventes_A_session (€) (m)'}, inplace=True),cas.rename(columns={'ventes_B (€) (m)':'ventes_B_session (€) (m)'}, inplace=True),cas.rename(columns={'ventes_C (€) (m)':'ventes_C_session (€) (m)'}, inplace=True),cas.rename(columns={'price':'ventes_session'}, inplace=True)

cac['fréquence d’achat mensuel (t)']=cac['nombre_sessions']/3/12
cac['Panier moyen (€)']=cac['ventes_client (€)']/cac['nombre_sessions']

cacc=cac
cacc['nombre_clients_somme']=1
cacc=cacc.groupby(['Age']).agg({'ventes_client (€)': np.sum, 'ventes_A (€)': np.sum, 'ventes_B (€)': np.sum, 'ventes_C (€)': np.sum, 'ventes_A (€) (f)': np.sum, 'ventes_B (€) (f)': np.sum, 'ventes_C (€) (f)': np.sum, 'ventes_A (€) (m)': np.sum,'ventes_B (€) (m)': np.sum, 'ventes_C (€) (m)': np.sum,'nombre_sessions': np.mean,'nombre_clients_somme': np.sum, 'fréquence d’achat mensuel (t)': np.mean, 'Panier moyen (€)': np.mean,'birth': np.mean,'Age': np.mean})
cacc.rename(columns={'nombre_sessions':'nombre_sessions_moyen'}, inplace=True)

jours=cas.groupby([cas['date'].dt.date]).sum()
mois=cas.set_index('date').groupby(pd.Grouper(freq='M')).sum()

year=cas.set_index('date').groupby(pd.Grouper(freq='Y')).sum()
jours[ '30jours_moyenne' ] = jours.ventes_session.rolling(30).mean()

jours[ '30jours_moyenne categ A' ] = jours['ventes_A_session (€)'].rolling(30).mean()
jours[ '30jours_moyenne categ B' ] = jours['ventes_B_session (€)'].rolling(30).mean()
jours[ '30jours_moyenne categ C' ] = jours['ventes_C_session (€)'].rolling(30).mean()

mois['6mois_moyenne']= mois.ventes_session.rolling(6).mean()
year.index = pd.to_datetime(year.index, format = '%m/%d/%Y').strftime('%Y')
year.rename(columns={'ventes_session':'Ventes an'}, inplace=True)
mois.index = pd.to_datetime(mois.index, format = '%Y/%m/%d').strftime('%m-%Y')
mois.rename(columns={'ventes_session':'Ventes mois'}, inplace=True)
mois.drop(columns={'Age','birth'}, inplace=True)
year.drop(columns={'Age','birth'}, inplace=True)

```
